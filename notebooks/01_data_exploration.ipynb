```python
import pandas as pd
```

```python
# Load India dataset
india_df = pd.read_csv("data/raw/INvideos.csv")

# Add a new column for country
india_df['country'] = 'India'

# Check first 5 rows
india_df.head()
```

```python
# Load US dataset
us_df = pd.read_csv("data/raw/USvideos.csv")

# Add a new column for country
us_df['country'] = 'United States'

# Check first 5 rows
us_df.head()
```

```python
# Combine India and US datasets
youtube_data = pd.concat([india_df, us_df], ignore_index=True)

# Check the shape (rows, columns)
print(youtube_data.shape)

# Preview the combined dataset
youtube_data.head()
```

```python
# Convert trending_date to proper datetime
youtube_data['trending_date'] = pd.to_datetime(
    youtube_data['trending_date'],
    format='%y.%d.%m'
)

youtube_data[['trending_date']].head()
```

```python
# Convert publish_time to datetime
youtube_data['publish_time'] = pd.to_datetime(youtube_data['publish_time'])

youtube_data[['publish_time']].head()
```

```python
# Sort by views so highest view count comes first
youtube_data = youtube_data.sort_values(by='views', ascending=False)

# Drop duplicate videos, keeping the one with max views
youtube_data = youtube_data.drop_duplicates(subset='video_id', keep='first')

# Reset index after dropping duplicates
youtube_data = youtube_data.reset_index(drop=True)

youtube_data[['video_id', 'views']].head()
```

```python
import json

# Load India category JSON
with open("data/raw/IN_category_id.json", "r", encoding="utf-8") as f:
    india_categories = json.load(f)
```

```python
india_categories.keys()
```

```python
india_categories['items'][:2]
```

```python
# Create category mapping dictionary for India
india_category_map = {
    int(item['id']): item['snippet']['title']
    for item in india_categories['items']
}

# Check mapping
india_category_map
```

```python
# Map category_id to category name
youtube_data['category_name'] = youtube_data['category_id'].map(india_category_map)

# Check result
youtube_data[['category_id', 'category_name']].head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply cleaning to titles
youtube_data['clean_title'] = youtube_data['title'].apply(clean_text)

# Check result
youtube_data[['title', 'clean_title']].head()
```

```python
# Combine all cleaned titles into one string
all_titles_text = " ".join(youtube_data['clean_title'])

all_titles_text[:200]
```

```python
from collections import Counter

# Split text into words
words = all_titles_text.split()

# Count word frequencies
word_freq = Counter(words)

# View top 20 most common words
word_freq.most_common(20)
```

```python
from nltk.corpus import stopwords
import nltk

# Download stopwords (only once)
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# Remove stopwords
filtered_words = [word for word in words if word not in stop_words]

# Recalculate frequencies
filtered_word_freq = Counter(filtered_words)

filtered_word_freq.most_common(20)
```

```python
import matplotlib.pyplot as plt

# Get top 10 words
top_words = filtered_word_freq.most_common(10)

# Separate words and counts
words_list = [word for word, count in top_words]
counts_list = [count for word, count in top_words]

# Create bar plot
plt.figure(figsize=(10, 5))
plt.bar(words_list, counts_list)
plt.xticks(rotation=45)
plt.title("Top Words in Viral YouTube Video Titles")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Select numeric engagement columns
engagement_cols = ['views', 'likes', 'comment_count']

# Compute correlation matrix
corr_matrix = youtube_data[engagement_cols].corr()

print(corr_matrix)
```

```python
# Count number of videos per category
top_categories = youtube_data['category_name'].value_counts().head(10)

print(top_categories)
```

```python
# Count categories by country
top_categories_country = youtube_data.groupby('country')['category_name'].value_counts().unstack().fillna(0)

top_categories_country
```

```python
top_categories_country.T.plot(kind='bar', figsize=(12,6), stacked=True)
plt.title("Top YouTube Categories by Country")
plt.xlabel("Category")
plt.ylabel("Number of Trending Videos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
